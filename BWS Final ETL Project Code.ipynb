```python
import pandas as pd
import numpy as np
import datetime
```

```python
#read csv
agriculture_df = pd.read_csv("FAO.csv",encoding ='latin1')
agriculture_df.head()
```

```python
# read globaltemp csv
temperature_df = pd.read_csv("globallandtemperaturesbycountry.csv")
temperature_df.head()
```

```python
# Extract year from datetime in order to create the mask
temperature_df["year"] = pd.DatetimeIndex(temperature_df['dt']).year
temperature_df.head()
```

```python
#set date range
start_date = '1991-12-01'
end_date = '2012-12-01'
```

```python
#created mask to filter by date range
mask = (temperature_df['dt']>start_date)&(temperature_df['dt']<=end_date)
```

```python
temperature_filtered_df = temperature_df.loc[mask]
temperature_filtered_df.head()
```

```python
#rename columns
agriculture_df.rename (columns={"Area":"Country"})
```

```python
# Create new dataframe by dropping unnecessary columns
new_agriculture_df = agriculture_df[['Area Abbreviation', 'Item Code', 'Item', "Y1992","Y1993","Y1994","Y1995","Y1996","Y1997","Y1998","Y1999","Y2000","Y2001","Y2002","Y2003","Y2004","Y2005","Y2006","Y2007","Y2008","Y2009","Y2010","Y2011","Y2012" ]].copy()
new_agriculture_df.head()
```

```python
# save dataframe to CSV
new_agriculture_df.to_csv("C:/Users/skinn/Desktop/ETL Project//new_agriculture_df.csv")
```

```python
# pd.Series(df.A.values,index=df.B).to_dict() to map key values to drop duplicates in a Dataframe
mapping_= pd.Series(agriculture_df['Area Abbreviation'].values,index=agriculture_df['Area']).to_dict()
mapping_
```

```python
#create dataframe from list of key valued pairs
mapping_df = pd.DataFrame.from_dict(mapping_, orient='index')
mapping_df.rename (columns={0:"Country Code"}, inplace=True)
mapping_df.head(50)
```

```python
#rename columns
new_agriculture_df.rename (columns={"Area Abbreviation":"Country Code"})
```

```python
# save dataframe to CSV file
mapping_df.to_csv("C:/Users/skinn/Desktop/ETL Project//mapping_df.csv")
```

```python
# used unique function to create a list of unique values 
country_df = agriculture_df.Area.unique()
country_df
```

```python
#used unique function to create a list of country codes
area_abb_df = agriculture_df["Area Abbreviation"].unique()
```

```python
#create Dataframe of unique country codes
new_area_abb_df = pd.DataFrame(area_abb_df)
new_area_abb_df.head()
```

```python
#renamed columns
new_area_abb_df.rename (columns={0:"Country Code"}, inplace=True)
new_area_abb_df.head()
```

```python
#read csv
temperature_combined = pd.read_csv("global_df.csv")
temperature_combined.head()
```

```python
#merged two data frames to add country code 
global_merge = mapping_df.merge(temperature_combined.set_index(['Country']), left_index=True, right_index=True, how = 'right')

global_merge.head()
```

```python
#dropped duplicates for clean dataframe
global_merge_df = global_merge.dropna(how='any')
global_merge_df.head()
```
