```python
csv_file = "GlobalLandTempByCountry.csv"
global_df = pd.read_csv(csv_file)
global_df.head()
```

```python
global_df['date']=pd.to_datetime(global_df['dt'])
```

```python
start_date = '1991-12-01'
end_date = '2012-12-01'
```

```python
mask = (global_df['dt'] > start_date) & (global_df['dt'] <= end_date)
```

```python
global_df = global_df.loc[mask]
global_df = global_df.drop(columns='dt')
global_df
```

```python
global_df['year'] = pd.DatetimeIndex(global_df['date']).year
global_df.head()

```

```python
countries_temp=global_df['Country'].unique()
```

```python
countries_temp = pd.DataFrame(countries_temp)
countries_temp
```

```python
global_df_list = global_df.groupby(['Country', 'year'])['AverageTemperature'].mean()
```

```python
global_df = pd.DataFrame(global_df_list)
global_df
```

```python
global_df.to_csv("global_df.csv")
```
